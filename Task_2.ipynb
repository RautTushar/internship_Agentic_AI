# **Problem** 1

```python
sentence = input("Enter sentence: ")
words_list = sentence.lower().split()
unique_words = set(words_list)
print(f"\nUnique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

# **Problem** 2

```python
# Problem 2
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_paid_employee = max(employees, key=employees.get)
salary = employees[highest_paid_employee]
print(f"Highest Salary: {highest_paid_employee} - {salary}")
```

# Problem **3**

```python
numbers = [45, 22, 89, 10, 66]
max_val = numbers[0]
min_val = numbers[0]
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print(f"List: {numbers}")
print(f"Max: {max_val}")
print(f"Min: {min_val}")
```

# Problem **4**

```python
prices = [450, 1200, 899, 1500, 300]
threshold = 1000
count = 0
for price in prices:
    if price > threshold:
        count += 1
print(f"Products above 1000: {count}")
```

# Problem **5**

```python
attendance = ["P", "P", "A", "P", "P"]
present_days = attendance.count("P")
total_days = len(attendance)
percentage = (present_days / total_days) * 100
print(f"Attendance Percentage: {percentage}")
```

# Problem **7**

```python
text = "pythonp"
frequency = {}
for char in text:
    frequency[char] = frequency.get(char, 0) + 1
print(frequency)
```
