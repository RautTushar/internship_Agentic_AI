# Problem **1**

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
max_score = max(employees.values())
top_performers = [name for name, score in employees.items() if score == max_score]
names_str = ", ".join(top_performers)
print(f"Top Performers Eligible for Bonus: {names_str} (Score: {max_score})")
```

# Problem **2**

```python
search_query = "Buy mobile phone buy phone online"
words = search_query.lower().split()
counts = {}
for word in words:
    counts[word] = counts.get(word, 0) + 1
repeated_keywords = {word: count for word, count in counts.items() if count > 1}

print(repeated_keywords)
```

# Problem **3**

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = [(hour, val) for hour, val in enumerate(sensor_readings) if val % 2 == 0]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# Problem **4**

```python
emails = ["ravi@gmail.com", "anita@yahoo.com", "kiran@gmail.com", "suresh@gmail.com", "meena@yahoo.com"]
domain_counts = {}
for email in emails:
    domain = email.split("@")[1]
    domain_counts[domain] = domain_counts.get(domain, 0) + 1
total_emails = len(emails)
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

# Problem **5**

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
average_sales = sum(sales) / len(sales)
threshold = average_sales * 1.30
for i, sale in enumerate(sales):
    if sale > threshold:
        print(f"Day {i + 1}: {sale}")
```
