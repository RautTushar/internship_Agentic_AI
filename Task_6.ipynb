# Problem **1**

```python
def parking_system(capacity, logs):
    cars_inside = 0
    peak_count = 0
    for entry in logs:
        if entry == "IN":
            cars_inside += 1
        elif entry == "OUT":
            if cars_inside > 0:
                cars_inside -= 1
        if cars_inside > peak_count:
            peak_count = cars_inside
    print("Currently Parked Vehicles:", cars_inside)
    if cars_inside > capacity:
        print("Parking Status: Full / Over Capacity")
    else:
        print("Parking Status: Available")
    print("Peak Parking Usage:", peak_count)
parking_system(50, ["IN","IN","IN","OUT","IN","IN","OUT"])
```

# Problem **2**

```python
def estimate_delivery_time(distance, traffic, weather):
    time_needed = distance * 6
    if traffic == "High":
        time_needed += 20
    elif traffic == "Medium":
        time_needed += 10
    else:
        time_needed += 5
    if weather == "Rainy":
        time_needed += 10
    elif weather == "Stormy":
        time_needed += 20
    print("Estimated Delivery Time:", time_needed, "minutes")
estimate_delivery_time(8, "High", "Rainy")
```

# Problem **3**

```python
def theatre_status(total_seats, booked_list):
    booked_seats = len(booked_list)
    percent = (booked_seats / total_seats) * 100
    print("Occupancy:", int(percent), "%")
    if percent == 100:
        print("Show Status: Housefull")
    elif percent >= 70:
        print("Show Status: Almost Full")
        print("Suggestion: Consider opening extra show")
    else:
        print("Show Status: Seats Available")
theatre_status(200, [1]*150)
```

# Problem **4**

```python
def check_server_load(cpu_values):
    total = 0
    for value in cpu_values:
        total += value
    average = total / len(cpu_values)
    print("Average CPU Load:", int(average), "%")
    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
check_server_load([45, 60, 70, 85, 90])
```

# Problem **5**

```python
def classroom_resources(usage_dict):
    overused_items = []
    for item in usage_dict:
        if usage_dict[item] > 8:
            overused_items.append(item)
    if len(overused_items) > 0:
        print("Overused Resources:", ", ".join(overused_items))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
classroom_resources({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

# Problem **6**

```python
def manage_registration(capacity, total_registrations):
    if total_registrations <= capacity:
        confirmed = total_registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = total_registrations - capacity
        status = "Closed"
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)
manage_registration(100, 105)
```
