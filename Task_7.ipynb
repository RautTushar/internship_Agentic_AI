# Problem **1**

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"
    def set_action(self, action):
        self.status = action.upper()
    def display_status(self):
        print(f"\n{self.name} is {self.status}")
name_in = input("Enter Light Name: ")
action_in = input("Enter Action (ON/OFF): ")
light = SmartLight(name_in)
light.set_action(action_in)
light.display_status()
```

# Problem **2**

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    def display_card(self):
        print("\n--- Employee ID Card ---")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")
name = input("Enter Employee Name: ")
eid = input("Enter Employee ID: ")
dept = input("Enter Department: ")
emp = Employee(name, eid, dept)
emp.display_card()
```

# Problem **3**

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def display_contact(self):
        print("\nContact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")
c_name = input("Enter Contact Name: ")
c_phone = input("Enter Phone Number: ")
contact = Contact(c_name, c_phone)
contact.display_contact()
```

# Problem **4**

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def print_tag(self):
        print(f"\nProduct: {self.name}")
        print(f"Price: ₹{self.price}")
p_name = input("Enter Product Name: ")
p_price = int(input("Enter Price: "))
item = Product(p_name, p_price)
item.print_tag()
```

# Problem **5**

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    def display_details(self):
        print(f"\nMovie: {self.name}")
        print(f"Rating: {self.rating}")
m_name = input("Enter Movie Name: ")
m_rating = float(input("Enter Rating: "))
film = Movie(m_name, m_rating)
film.display_details()
```

# Problem **6**

```python
class DeliveryManager:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address
    def print_details(self):
        print("\n--- Delivery Details ---")
        print(f"Customer: {self.customer}")
        print(f"Address: {self.address}")
cust = input("Enter Customer Name: ")
addr = input("Enter Address: ")
delivery = DeliveryManager(cust, addr)
delivery.print_details()
```
