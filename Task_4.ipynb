# Problem **1**

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    while True:
        try:
            amount = int(input("Enter recharge amount: "))
            if amount < 50:
                print("Minimum recharge amount is ₹50. Try again.")
                continue

            if amount in valid_plans:
                print(f"Recharge of ₹{amount} successful!")
                break
            else:
                print(f"Invalid plan. Choose from {valid_plans}.")
        except ValueError:
            print("Please enter a valid numeric amount.")
```

# Problem **2**

```python
def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(f"Alert: {item} is low on stock ({stock} units). Reorder now!")
        else:
            print(f"{item}: Stock OK ({stock} units).")

inventory = {"Laptops": 20, "Mice": 10, "Keyboards": 5, "Monitors": 15}
check_inventory(inventory)
```

# Problem **3**

```python
def process_student_result(marks):
    if not marks:
        return "No marks provided"

    total = sum(marks)
    average = total / len(marks)

    status = "Pass" if average >= 50 else "Fail"
    print(f"Average Marks: {average:.2f} | Status: {status}")
    return status

student_marks = [55, 60, 45, 70, 30]
process_student_result(student_marks)
```

# Problem **4**

```python
def calculate_fare(km, is_peak):
    base_fare = 50
    rate_per_km = 12
    total = base_fare + (km * rate_per_km)

    if is_peak:
        total *= 1.25  # 25% extra
    return total

while True:
    try:
        dist = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ").lower() == 'yes'

        fare = calculate_fare(dist, peak)
        print(f"Estimated Fare: ₹{fare:.2f}")

        retry = input("Calculate another fare? (y/n): ").lower()
        if retry != 'y':
            break
    except ValueError:
        print("Invalid input. Please enter numbers for distance.")
```

# Problem **5**

```python
def check_attendance_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = attendance_list.count("P")

    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        return f"Eligible ({percentage:.1f}%)"
    else:
        return f"Not Eligible ({percentage:.1f}%)"

attendance = ["P", "P", "A", "P", "P", "A", "P"]
print(check_attendance_eligibility(attendance))
```

# Problem **6**

```python
def check_password_strength(password):
    if len(password) < 8:
        return "Weak: Minimum 8 characters required."

    has_digit = any(char.isdigit() for char in password)
    has_special = any(char in "@#$" for char in password)

    if has_digit and has_special:
        return "Strong Password"
    elif has_digit or has_special:
        return "Medium Password: Add both digits and special characters (@#$) for full strength."
    else:
        return "Weak: Missing digits and special characters."

# Example usage
print(check_password_strength("Pass123@"))
```
