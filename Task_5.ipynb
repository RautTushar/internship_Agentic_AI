# Problem **1**

```python
def analyze_engagement(likes_list):
    total_likes = 0
    for likes in likes_list:
        total_likes += likes
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")
post_likes = [400, 450, 300]
analyze_engagement(post_likes)
```

```python
def monitor_medicine_stock(current_stock):
    if current_stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print(f"Medicine Stock: {current_stock}")
    print(f"Status: {status}")

monitor_medicine_stock(6)
```

# Problem **3**

```python
def check_rainfall_adequacy(rainfall_data, required_level):
    total = 0
    for data in rainfall_data:
        total += data
    average = total / len(rainfall_data)
    status = "Adequate Rainfall" if average >= required_level else "Inadequate Rainfall"
    print(f"Average Rainfall: {average:.0f}")
    print(f"Rainfall Status: {status}")
check_rainfall_adequacy([60, 80, 70, 78], 70)
```

# Problem **4**

```python
def detect_duplicate_accounts(usernames):
    # A set removes all duplicates automatically
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"
    print(f"Duplicate Accounts Found: {result}")
detect_duplicate_accounts(["user1", "admin", "user1", "guest"])
```

# Problem **7**

```python
def check_app_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")
check_app_health(7)
```

# Problem **9**

```python
def check_exam_eligibility(attendance_record):
    total_days = len(attendance_record)
    present_days = attendance_record.count("P")
    percentage = (present_days / total_days) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"
    print(f"Attendance Percentage: {percentage:.1f}")
    print(f"Exam Eligibility: {status}")
check_exam_eligibility(["P", "P", "P", "A", "P"])
```

# Problem **10**

```python
def analyze_electricity_bill(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    if bill < 500:
        usage_type = "Low Usage"
    elif 500 <= bill <= 1500:
        usage_type = "Moderate Usage"
    else:
        usage_type = "High Usage"
    return bill, usage_type
total_bill, level = analyze_electricity_bill(250)
print(f"Total Bill: ₹{total_bill}")
print(f"Usage Level: {level}")
```
